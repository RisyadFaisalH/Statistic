**THE DATA FROM THE SOURCE**


```
data <- read.csv('https://github.com/RisyadFaisalH/Statistic/blob/main/Statistic.csv?raw=true')

head(data)

#Condition where last 2 column is deleted
```

**PIE CHART**

```
#Pie chart for gender
Gender <- as.data.frame(table(data$Gender))
prct <- round(100*Gender$Freq/sum(Gender$Freq))

pie(Gender$Freq, labels = paste(Gender$Var1, sep = " ", prct, "%"))

#Pie chart for Eduacational Level
EduLevel <- as.data.frame(table(data$Education_Level))
prct <- round(100*EduLevel$Freq/sum(EduLevel$Freq))

pie(EduLevel$Freq, labels = paste(EduLevel$Var1, sep = " ", prct, "%"))

#Pie chart for Income Category
inccat <- as.data.frame(table(data$Income_Category))
prct <- round(100*inccat$Freq/sum(inccat$Freq))

pie(inccat$Freq, labels = paste(inccat$Var1, sep = " ", prct, "%"))

#Pie chart for Marital_Status
mrtlsts <- as.data.frame(table(data$Marital_Status))
prct <- round(100*mrtlsts$Freq/sum(mrtlsts$Freq))

pie(mrtlsts$Freq, labels = paste(mrtlsts$Var1, sep = " ", prct, "%"))
```

**HISTOGRAM AND BOXPLOT FOR CUSTOMER AGE**

```
x <- data.frame(data$Customer_Age)
head(x)
hist(data$Customer_Age, xlab= "Customer Age", main = "Customer Age", col = "blue")
boxplot(data$Customer_Age, notch = TRUE, varwidth = TRUE , horizontal = TRUE, main = "Boxplot Customer Age", col = "Red")
summary(data$Customer_Age)



```

**BOXPLOT FOR COMPARISON**

```
y <- data.frame(data$Attrition_Flag, data$Total_Relationship_Count)
head(y)

a <- data[data$Attrition_Flag == "Existing Customer", ]
b <- data[data$Attrition_Flag == "Attrited Customer", ]

boxplot(a$Total_Relationship_Count , b$Total_Relationship_Count, names = c("Existing customer", "Attrited Customer"), ylab = "Total Relationship Count", 
  main = "Comparison Between Existing and Attrited customer", col = c("Red", "Green"))

summary(a$Total_Relationship_Count)
summary(b$Total_Relationship_Count)
```

```
z <- data.frame(data$Dependent_count, data$Card_Category)
head(z)

x <- barplot(table(data$Dependent_count, data$Card_Category),beside = TRUE, main = "How many user per card category", col =c("red", "Green", "Blue", "White"))
text(x = x, y = data$Dependent_count, label = data$Dependent_count, pos = 1, cex = 0.9)
```

**SCATTER PLOT AND CORALATION**

```
v <- data.frame(data$Credit_Limit, data$Avg_Open_To_Buy)
head(v)

plot(data$Credit_Limit, data$Avg_Open_To_Buy, xlab = "Credit Limit", ylab = "Average Open to Buy", main = "Correlation between credit limit average open to buy")
cor(data$Credit_Limit, data$Avg_Open_To_Buy)

#signifincae level 5% = 0.05 --> conficence level 95%

cor.test(data$Credit_Limit, data$Avg_Open_To_Buy, conf.level = 0.95)
```
